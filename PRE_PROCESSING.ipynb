#### 1] Description :
- Collecting the headline as it is from the webpage
- Scraping the data from <H1 & H2> HTML tags respectively

```python
def get_description(h1,h2):
    h = h1 + '.' + h2                                
    comma = ','
    no_punct = ""
    for char in h:
        if char not in comma:
            no_punct = no_punct + char
    return (no_punct)
```

#### Removing Punctuation Marks

```python
def no_punctuations(description):
    punctuations = '''!â€™()-[]{};:'"\,<>./?@#$%^&*_~'''
    no_punct = ""
    for char in description:
        if char not in punctuations:
            no_punct = no_punct + char
    des = no_punct.split()
    return (des)
```

## <center>HTML SCRAPER</center>
- Single function which consists of all the above functions to extract information from webpages & storing everything inside a csv file

```python
def html_scraper(x):
    
    DESCRIPTION = [];DATE = [];MONTH = [];YEAR = [];CITY = [];STATE = [];ZONE = [];CASUALITIES = [];NUMBER_OF_CASUALITIES = [];
    EQUIPMENTS = [];LITERACY = [];POPULATION = [];NUMBER_OF_INCIDENTS = [];GROWTH_IN_GDP = [];INFLATION = [];TERRORIST_GROUP = [];
    
    response = uReq(x)
    page_soup = soup(response,'html.parser')
    h1 = page_soup.h1.text
    try:
        h2 = page_soup.h2.text
    except AttributeError as e:
        h2 = '---'
    
    description = get_description(h1,h2)                                            # PARAMETER - 1
    DESCRIPTION.append(description)
    div = page_soup.findAll('div')
    
    date = get_date(div,page_soup)    
    
    month = get_month(date)                                                         # PARAMETER - 2
    MONTH.append(month)
    
    year = get_year(date)                                                           # PARAMETER - 3
    YEAR.append(year)
    
    date = str(date)                                                                # PARAMETER - 4
    DATE.append(date)
    
    (city,state) = get_city_and_state(description)                                  # PARAMETER - 5 & 6
    CITY.append(city)
    STATE.append(state)
    
    zone = get_zone(state)                                                          # PARAMETER - 7
    ZONE.append(zone)
    
    equipments = get_equip(description)                                             # PARAMETER - 8
    EQUIPMENTS.append(equipments)
    
    literacy = get_literacy(state)                                                  # PARAMETER - 9
    LITERACY.append(literacy)
    
    population = get_pop(city)                                                      # PARAMETER - 10
    POPULATION.append(population)
    
    incidents = get_no_incidents(year)                                              # PARAMETER - 11
    NUMBER_OF_INCIDENTS.append(incidents)
    
    GDP = get_gdp_growth(year)                                                      # PARAMETER - 12
    GROWTH_IN_GDP.append(GDP)
    
    inflation = get_inflation(year)                                                 # PARAMETER - 13
    INFLATION.append(inflation)
    
    try:
        
        (casualities,number) = get_casualities(description)                             # PARAMETER - 14 & 15
        CASUALITIES.append(casualities)
        NUMBER_OF_CASUALITIES.append(number)
    except TypeError as e:
        
        CASUALITIES.append('No')
        NUMBER_OF_CASUALITIES.append('0')
    
    terrorist_group = get_terrorist_group(description)
    TERRORIST_GROUP.append(terrorist_group)                                         # PARAMETER - 16
    
    database = pd.DataFrame(columns = ['DESCRIPTION','DATE','MONTH','YEAR','CITY','STATE','ZONE','CASUALTIES','NUMBER OF CASUALTIES','EQUIPMENTS','LITERACY','POPULATION','NUMBER OF INCIDENTS','GROWTH IN GDP','INFLATION','TERRORIST GROUP'])
    database['DESCRIPTION'] = DESCRIPTION;database['DATE'] = DATE;database['MONTH'] = MONTH;database['YEAR'] = YEAR
    database['CITY'] = CITY;database['STATE'] = STATE;database['ZONE'] = ZONE;database['CASUALTIES'] = CASUALITIES
    database['NUMBER OF CASUALTIES'] = NUMBER_OF_CASUALITIES;database['EQUIPMENTS'] = EQUIPMENTS;database['LITERACY'] = LITERACY
    database['POPULATION'] = POPULATION;database['NUMBER OF INCIDENTS'] = NUMBER_OF_INCIDENTS;database['GROWTH IN GDP'] = GROWTH_IN_GDP
    database['INFLATION'] = INFLATION;database['TERRORIST GROUP'] = TERRORIST_GROUP
    
    return (database)
```
